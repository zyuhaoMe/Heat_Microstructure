```python
# import libaray
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
sns.set(rc={"figure.dpi":1600, 'savefig.dpi':1600})
sns.set_theme(style = "white")
# show all rows
pd.set_option('display.max_rows',None)
```

```python
# list all the data in the path
path = "./tecplot_points/"
pathNameList = os.listdir(path)
pathNameList
```

```python
# mannual give the name of the case
caseName = "stationaryAna_0";
```

```python
# find the list of the fileName with caseName
pathNameListThisCase = []
for i in range(len(pathNameList)):
    if caseName in pathNameList[i]:
        pathNameListThisCase.append(pathNameList[i])
        
pathNameListThisCase
```

```python
# read the thermal history of a point
f = open("./tecplot_points/"+pathNameListThisCase[1], "r")
tempListThisCase = (f.read()).split()
tempListThisCase = list(map(float, tempListThisCase))
numTimeStep = int(tempListThisCase[3])

# plot 
plt.plot(tempListThisCase[4:3 + numTimeStep], tempListThisCase[4 + numTimeStep: len(tempListThisCase) - 1])

# label 
plt.xlabel("Time(s)")
plt.ylabel("Temperature(K)")
```
